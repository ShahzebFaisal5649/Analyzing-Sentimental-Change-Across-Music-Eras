## Preprocess lyrics

```python
import re
def preprocess_lyrics(lyrics):
    processed_lyrics = re.sub(r'\d+\s*ContributorsTranslations\w*', '', lyrics)
    
    processed_lyrics = re.sub(r'\[.*?\]', '', processed_lyrics)
    processed_lyrics = re.sub(r'\(.*?\)', '', processed_lyrics)
    processed_lyrics = processed_lyrics.replace('\n', ' ')
    processed_lyrics = processed_lyrics.replace('contributors', ' ')
    processed_lyrics = processed_lyrics.replace('Lyrics', ' ')
    processed_lyrics = processed_lyrics.replace('[0-9]+', '')
    return processed_lyrics

```

```python
def textdata_cleaner(data):
    data = data.str.replace("[\(\[].*?[\)\]]", '')
    data = data.str.replace("\n", ' ')
    data = data.str.lower()
    data = data.str.replace('[{}]'.format(string.punctuation), '')
    
    return data
def remove_stopwords(text):    # text is a list/series of string to clean
    clean_text = list()
    
    words=["yeah","ya","na","wan","uh","gon","ima","mm","uhhuh","bout","em","nigga","niggas","got","ta","lil","ol","hey",
          "oooh","ooh","oh","youre","dont","im","youve","ive","theres","ill","yaka","lalalala","la","da","di","yuh",
          "shawty","oohooh","shoorah","mmmmmm","ook","bidibambambambam","shh","bro","ho","aint","cant","know","bambam",
          "shitll","tonka"]
    ignore= (stopwords.words('english') + words)
    
    for i in text:
        words = nltk.word_tokenize(i)
        for element in ignore:
            words = list(filter(lambda x: x!= element and len(x) > 1, words))
        lyric = " ".join(words)
        clean_text.append(lyric)
    
    return clean_text

def convert_duration(duration): #time into total seconds
    minutes, seconds = map(int, duration.split(':'))
    return minutes * 60 + seconds
```

```python
merged_df['Duration_in_seconds'] = merged_df['Duration_in_seconds'].apply(convert_duration)
```
