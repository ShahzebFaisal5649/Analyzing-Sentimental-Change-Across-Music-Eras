```python
def convert_miliseconds_to_seconds(duration_ms):
    seconds = duration_ms / 1000
    minutes, seconds = divmod(seconds, 60)
    return f"{int(minutes)}:{int(seconds):02d}"
```

# Get ALL TRACKS FROM SPOTIFY PLAYLIST DATA

```python
df_playlist=pd.read_csv('Spotify_playlist_data.csv')
```

```python
def get_all_tracks(playlist_uri):
    tracks = []
    retries = 3  # Number of retries
    for attempt in range(retries):
        try:
            results = SP.playlist_tracks(playlist_uri)
            while results['items']:
                for track in results['items']:
                    track_uri = track["track"]["uri"]
                    track_name = track["track"]["name"]
                    artist_name = track["track"]["artists"][0]["name"]
                    release_date = SP.track(track_uri)['album']['release_date']
                    duration_ms = track["track"]["duration_ms"]
                    duration_in_seconds = convert_miliseconds_to_seconds(duration_ms)
                    year = release_date[:4]
                    result = {"Song_title": track_name, "Release_year": year,"Duration_in_seconds":duration_in_seconds,
                              "artist_name":artist_name}
                    tracks.append(result)

                # Check if there are more tracks to retrieve
                if results['next']:
                    results = SP.next(results)
                else:
                    break
            return tracks
        except ReadTimeout as e:
            if attempt < retries - 1:
                print(f"Request timed out. Retrying... (Attempt {attempt + 1}/{retries})")
                sleep(2 ** attempt)  # Exponential backoff
            else:
                raise e


```

## Store all songs in csv:

```python
columns_tracks = ['Playlist_uri', 'Song_title', 'Release_year', 'Duration_in_seconds', 'artist_name']
df_tracks = pd.DataFrame(columns=columns_tracks)

for index, row in df_playlist.iloc[80:110].iterrows():
    playlist_uri = row['Playlist_uri']
    print(index)
    tracks = get_all_tracks(playlist_uri)
    if(tracks):
        tracks_data = pd.DataFrame(tracks)
        tracks_data['Playlist_uri'] = playlist_uri
    
        df_tracks = df_tracks.append(tracks_data, ignore_index=True)
    else:
        print("Lyrics not found")
```
